**Importing all Necessary Libraries**

```python
import requests
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import json
%matplotlib inline
```

**Fetching dummy Data**

```python
url = "https://dummyjson.com/users?limit=10"
print(f"Fetching data from {url}...")
response = requests.get(url, timeout=10)
data = response.json()['users']
print("Data Loaded")
```

```python
data
```

**Extracting Name and Score (using 'weight' as proxy for score)**

```python
students = [f"{u['firstName']} {u['lastName']}" for u in data]
scores = [u['weight'] for u in data]
avg_score = sum(scores) / len(scores)
```

**Visualization**


```python
plt.figure(figsize=(10, 5))
bars = plt.bar(students, scores, color='skyblue', edgecolor='black')
        
 # Add average line
plt.axhline(y=avg_score, color='r', linestyle='--', label=f'Avg: {avg_score:.1f}')
        
plt.title('Student Test Scores (Fetched from API)')
plt.xlabel('Student Name')
plt.ylabel('Score')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
        
# Show plot inline
plt.show()
```
